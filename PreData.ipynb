```python
# Hàm kiểm tra câu hợp lệ
def is_valid_sentence(s, lang='en'):
    if not isinstance(s, str) or len(s.strip().split()) < 2 or len(s.strip().split()) > 20:
        return False
    if re.search(r'<.*?>', s):  # loại tag HTML
        return False
    if re.search(r'[^\w\s\.\,\?\!à-ỹÀ-Ỹ]', s):  # ký tự lạ
        return False
    if lang == 'en' and re.search(r'\b(fuck|shit|damn|bitch|fucking)\b', s, re.IGNORECASE):
        return False
    if lang == 'vi' and re.search(r'\b(đm|vl|vkl|cl|cmm)\b', s, re.IGNORECASE):
        return False
    if lang == 'vi':
        for w in s.split():
            if w.count > 6: return False
    return True
```

```python
# Làm sạch chuỗi
def clean_text(text):
    text = re.sub(r'\s+', ' ', text.strip())  # bỏ khoảng trắng thừa
    text = re.sub(r'[^\w\s\.\,\?\!à-ỹÀ-Ỹ]', '', text)  # bỏ ký tự không hợp lệ
    text = re.sub(
        r"[^a-zA-Z0-9\s\.,!?àáảãạăằắẳẵặâầấẩẫậèéẻẽẹêềếểễệìíỉĩịòóỏõọôồốổỗộơờớởỡợùúủũụưừứửữựỳýỷỹỵđÀÁẢÃẠĂẰẮẲẴẶÂẦẤẨẪẬÈÉẺẼẸÊỀẾỂỄỆÌÍỈĨỊÒÓỎÕỌÔỒỐỔỖỘƠỜỚỞỠỢÙÚỦŨỤƯỪỨỬỮỰỲÝỶỸỴĐ]",
        '', text
    )
    # Danh sách thay thế không trang trọng → trung tính/lịch sự (ưu tiên cụm dài trước)
    # replacements = {
    #     r'\btụi mày\b': 'các bạn',
    #     r'\bmày\b': 'bạn',
    #     r'\btao\b': 'tôi',
    #     r'\bhắn\b': 'anh ấy'
    # }

    # Thực hiện thay thế an toàn bằng regex (theo từ vựng, có \b giới hạn từ)
    for pattern, replacement in replacements.items():
        text = re.sub(pattern, replacement, text, flags=re.IGNORECASE)

    return text.lower()

# Tiền xử lý từng dòng
def preprocess_pair(en, vi):
    en = clean_text(en)
    vi = clean_text(vi)
    return en, vi
```

```python
# Áp dụng lọc và xử lý
cleaned_data = []
for en, vi in zip(df['en'], df['vi']):
    if is_valid_sentence(en, lang='en') and is_valid_sentence(vi, lang='vi'):
        en, vi = preprocess_pair(en, vi)
        if 0.5 <= len(vi.split()) / len(en.split()) <= 2:  # độ dài tương đối hợp lý
            cleaned_data.append((en, vi))

# Tạo DataFrame kết quả và loại trùng lặp
clean_df = pd.DataFrame(cleaned_data, columns=['en', 'vi']).drop_duplicates()

clean_df.head()
```

```python
file_path = "./Corpus/En_Vi.tsv"
df_raw = pd.read_csv(file_path, sep='\t', header=None, names=['id_en', 'en', 'id_vi', 'vi'], encoding='utf-8')
```

```python
cleaned_data = []
for en, vi in zip(df_raw['en'], df_raw['vi']):
    if is_valid_sentence(en, lang='en') and is_valid_sentence(vi, lang='vi'):
        en_clean, vi_clean = preprocess_pair(en, vi)
        if len(en_clean.split()) > 0 and 0.5 <= len(vi_clean.split()) / len(en_clean.split()) <= 2:
            cleaned_data.append((en_clean, vi_clean))

# Tạo DataFrame kết quả và loại trùng lặp
clean_df = pd.DataFrame(cleaned_data, columns=['en', 'vi']).drop_duplicates()

# Xuất ra CSV
output_path = "En_Vi_Tatoeba.csv"
clean_df.to_csv(output_path, index=False, encoding='utf-8')

output_path
```
